#                                   KNN CLASSIFIER

```python
import numpy as np
import matplotlib.pyplot as plt
import KNeighborsClassifier as KNN
```

```python
data=np.loadtxt(open("C:\\Users\\Rishabh\\Desktop\\Machine Learning\\KNN Classifier\\Outlier.txt","r"),delimiter=',')
X=data[:,0:-1]
y=data[:,-1:].flatten()
m, n = X.shape
print("shape of data:",m,n)
```

```python
#defining object of KNeighborClassifier and shuffling data
clf=KNN.KNeighborsClassifier()
clf.shuffle_in_unison(X,y)
```

```python
#normalizing features
X=clf.normalize(X)
```

```python
#dividing into training data and testing data in the ratio - 70% : 30%

[a,b]=np.shape(X)
X_train=X[0:(7*a)//10,:]
y_train=y[0:(7*a)//10]
X_test=X[(7*a)//10:a,:]
y_test=y[(7*a)//10:a]
```

------------------------------------------Visualising Training Data------------------------------------------------------------

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
plt.xlabel("training dataset 1")
plt.ylabel("training dataset 2")
plt.title("Training Data")
plt.show()
```

```python
clf.fit(X_train,y_train)
```

------------------------------------------Visualising Predicted Data------------------------------------------------------------

```python
y_pred=clf.predict(X_test)
plt.scatter(X_test[:,0],X_test[:,1],c=y_test)
plt.xlabel("Testing dataset 1")
plt.ylabel("Testing dataset 2")
plt.title("Predicted Data")
plt.show()
```

  -------------------------------------------Some Predictions---------------------------------------------------

```python
indices=np.random.choice(m,10)
print("Test set:      ",y[indices])
print("My predictions:",clf.predict(X[indices,:]))
```

-------------------------------------Calculating accuracy--------------------------------------------------------------------

```python
#calculating accuracy of the model on test data
Accuracy=clf.accuracy(X_test,y_test)
print("accuracy of my model= %0.1f" %(Accuracy),"%")
```
