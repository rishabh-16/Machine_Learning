#                                   K Means Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
import K_means_Clustering as KMeans
```

```python
data=np.loadtxt(open("Image1.txt","r"),delimiter=',')
X=data[:,0:-1]
m, n = X.shape
print("shape of data:",m,n)
```

```python
#defining object of KMeans and shuffling data
cluster=KMeans.KMeans()
```

```python
#normalizing features
X=cluster.normalize(X)
```

```python
cluster.fit(X)
cl=cluster.assign_clusters(100,2)
```

------------------------------------------Visualising Training Data------------------------------------------------------------

```python
plt.scatter(X[:,0],X[:,1])
plt.xlabel("training dataset 1")
plt.ylabel("training dataset 2")
plt.title("Training Data")
plt.show()
```

------------------------------------------Visualising Predicted Data------------------------------------------------------------

```python

plt.scatter(X[:,0],X[:,1],c=cl)
plt.xlabel("Testing dataset 1")
plt.ylabel("Testing dataset 2")
plt.title("Predicted Data")
plt.show()
```

  -------------------------------------------Some Predictions---------------------------------------------------

```python
indices=np.random.choice(range(m),10)
c_pred=cluster.predict_clusters(X[indices,:])
print(np.concatenate((X[indices,:],c_pred[:,None]),axis=1))
```
